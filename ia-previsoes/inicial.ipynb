# Projeto Python IA: Inteligência Artificial e Previsões

### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

Arquivos da aula: https://drive.google.com/drive/folders/1FbDqVq4XLvU85VBlVIMJ73p9oOu6u2-J?usp=drive_link

```python
# Passo a Passo
# Passo 0 - Entender a empresa e o desafio da empresa 
# Passo 1 - Importa a base de dados e analisar
import pandas as pd

dados = pd.read_csv('C:/Users/Sabrina Gama/Documents/jornada-python/ia-previsoes/docs/clientes.csv')
display(dados.head(10))

```

```python
# Passo 2 - Preparar a base de dados para a Inteligência Artificial (marchine Learning)

# Já sei meu ALVO - SCORE_CREDITO

# Mostrar todas as colunas

print('Verificar se tem valores vazios')
display(dados.isnull().sum())

print('A descrição dos dados')
display(dados.describe())

pd.set_option('display.max_rows', None)
print('Informações dos dtypes dos dados')
display(dados.info())

# Vou precisar fazer LABELENCODER (trocar object por int ou float sem perder a informação)
from sklearn.preprocessing import LabelEncoder

# profissão
    # cientista - 1
    # bombeiro - 2
    # engenheiro - 3
    # dentista - 4
    # artista - 5

codificador_profissao = LabelEncoder()
# minha coluna profissao vai ser minha - antiga coluna profissao - com a transfomação do codificador labelencoder
dados['profissao'] = codificador_profissao.fit_transform(dados['profissao'])

# mix_credito
codificador_credito = LabelEncoder()
dados['mix_credito'] = codificador_credito.fit_transform(dados['mix_credito'])

# comportamento_pagamento
codificador_comportamento = LabelEncoder()
dados['comportamento_pagamento'] = codificador_comportamento.fit_transform(dados['comportamento_pagamento'])

# se eu olha de novo as info dos meu dados, vou perceber que antes os dtypes que era object agora são int
display(dados.info())
```

```python
# Vamos precisar fazer duas divisao - (treinar para aprender) e teste
#   Quem é a coluna que vai usar para fazer a previsao, chamamos de Y ou ALVO

# y - É a coluna da base de dados que eu quero PREVER
y = dados['score_credito']

# X - São as colunas da base de dados que eu vou usar para FAZER A PREVISAO
# Como eu quero usar toda a coluna, menos a coluna ALVO, eu uso o drop para tirar a coluna ALVO
X = dados.drop(columns=['score_credito', 'id_cliente'])

# SEPARAR DADOS DE TREINO E TESTE
from sklearn.model_selection import train_test_split
# por padrao o train test separa 75%, 25% - mas eu posso usar test_size=0.3
# recomendado começar com o padrão 75, 25. Depois faz algumas alterações
x_treino, x_teste, y_treino, y_teste = train_test_split(X, y)
```

```python
# Passo 3 - Treinar a Inteligência Artifical
#  Criar modelo: Nota do cliente: Boa, Ok, Ruim

# Vou importar dois modelos de inteligencia artificial para ver qual é o melhor modelo. 
# Qual IA?
    #  ARvore de Decisão -> RandomForest
    # Nearest Neighbors -> KNN -> Vizinhos proximos

# Importar a IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Criar a IA
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# Treinar a IA
# fit - treinar
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Passo 4 - Escolher o melhor modelo, do passo 3
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# Acurácia - 'Previsao'
from sklearn.metrics import accuracy_score
display(accuracy_score(y_teste, previsao_arvoredecisao))
display(accuracy_score(y_teste, previsao_knn))
print('Tem que me dar a porcentagem de acertividade de cada modelo que trabalhei (Arvore Decisao) e (Nearest Neighbors - KNN)')
```

```python
# Passo 5 - Usar o melhor modelo para fazer a previsão de novos clientes
# Vendo o passe 4, Visualizei que o modelo ARVORE DE DECISAO está mais preciso do que o KNN

# Importar novos clientes para fazer a previsao
dados_novos_clientes = pd.read_csv('C:/Users/Sabrina Gama/Documents/jornada-python/ia-previsoes/docs/novos_clientes.csv')

dados_novos_clientes['profissao'] = codificador_profissao.transform(dados_novos_clientes['profissao'])
dados_novos_clientes['mix_credito'] = codificador_credito.transform(dados_novos_clientes['mix_credito'])
dados_novos_clientes['comportamento_pagamento'] = codificador_comportamento.transform(dados_novos_clientes['comportamento_pagamento'])

display(dados_novos_clientes)
display(dados_novos_clientes.info())

nova_previsao = modelo_arvoredecisao.predict(dados_novos_clientes)
display(nova_previsao)
```
